**Traduzindo o dataframe:**

Traduzindo os nomes das colunas:

```python
df.rename(columns={'Year_Birth': 'ano_nascimento',
                   'Education':'educacao',
                   'Marital_Status':'status_conjugal',
                   'Income':'renda','Kidhome':'criancas_casa',
                   'Teenhome':'adolescentes_casa',
                   'Dt_Customer':'data_cliente',
                   'Recency':'recencia',
                   'MntWines':'vinhos',
                   'MntFruits':'frutas',
                   'MntMeatProducts':'carne_produtos',
                   'MntFishProducts':'peixe_produtos',
                   'MntSweetProducts':'doces_produtos',
                   'MntGoldProds':'ouro_produtos',
                   'NumDealsPurchases':'qtd_oferta_compras',
                   'NumWebPurchases':'qtd_web_compras',
                   'NumCatalogPurchases':'qtd_produtos_catalogo',
                   'NumStorePurchases':'qtd_produtos_armazenados',
                   'NumWebVisitsMonth':'qtd_web_visitantes_mensal',
                   'AcceptedCmp3':'cmp_aceita3',
                   'AcceptedCmp4':'cmp_aceita4',
                   'AcceptedCmp5':'cmp_aceita5',
                   'AcceptedCmp1':'cmp_aceita1',
                   'AcceptedCmp2':'cmp_aceita2',
                   'Complain':'reclamacoes',
                   'Z_CostContact':'z_contato_constante',
                   'Z_Revenue':'z_receita',
                   'Response':'resposta'
                   },inplace=True) 
```

```python
df
```

Traduzindo as linhas do df:

```python
print(df['educacao'].unique()) #coluna educacao: verificando os valores únicos para a tradução
```

```python
df.loc[df.educacao == 'Graduation',['educacao']] = 'graduacao'
df.loc[df.educacao == 'PhD',['educacao']] = 'doutorado'
df.loc[df.educacao == 'Master',['educacao']] = 'mestre'
df.loc[df.educacao == 'Basic',['educacao']] = 'basico'
df.loc[df.educacao == '2n Cycle',['educacao']] = 'segundo ciclo'
```

```python
print(df['educacao'].unique())
```

```python
print(df['status_conjugal'].unique()) #coluna status_conjugal: verificando os valores únicos para a tradução -> Dados para limpeza: alone, absurd, together e YOLO
```

```python
df.loc[df.status_conjugal == 'Single',['status_conjugal']] = 'Solteiro'
df.loc[df.status_conjugal == 'Married',['status_conjugal']] = 'Casado'
df.loc[df.status_conjugal == 'Divorced',['status_conjugal']] = 'Divorciado'
df.loc[df.status_conjugal == 'Widow',['status_conjugal']] = 'Viuvo'
```

```python
print(df['status_conjugal'].unique())
```

**Limpando os dados: ** deletando as colunas recencia, cmp_aceita3,	cmp_aceita4	cmp_aceita5,	cmp_aceita1,	cmp_aceita2, z_contato_constante,	z_receita	resposta
**Não estou vendo utilidade para elas nas análises!**

```python
df = df.drop(df.columns[[8, 20, 21, 22, 23, 24, 26, 27, 28]], axis=1)
```

```python
df
```

**Realizando a limpeza dos dados inconsistentes: Alone, Absurd, YOLO, Together.**

```python
df.loc[df.status_conjugal == 'Alone',['status_conjugal']] = None
df.loc[df.status_conjugal == 'Absurd',['status_conjugal']] = None
df.loc[df.status_conjugal == 'YOLO',['status_conjugal']] = None
df.loc[df.status_conjugal == 'Together',['status_conjugal']] = None
```

Verificando quantidade de valores nulos encontrados por coluna
